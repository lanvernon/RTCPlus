```python
def CollisionRate(filename, case):
    
    x = []
    y = []

    x2 = []
    y2 = []

    x3 = []
    y3 = []
    
    labels = ["RTC+", r"$\Delta_x$ = 50ms", r"$\Delta_x$ = 100ms"]
    x = np.arange(len(labels))
    width = 0.25

    with open('Data/'+ filename+'_CR_RTC.csv','r') as csvfile:
        lines = csv.reader(csvfile, delimiter=',')
        for row in lines:
            y.append(float(row[1])) 
    with open('Data/'+ filename+'_CR_50.csv','r') as csvfile:
        lines = csv.reader(csvfile, delimiter=',')
        for row in lines:
            y2.append(float(row[1])) 
    
    with open('Data/'+ filename+'_CR_100.csv','r') as csvfile:
        lines = csv.reader(csvfile, delimiter=',')
        for row in lines:
            y3.append(float(row[1])) 
            
    plt.bar(x + width, y, width) 
    plt.bar(x + 2*width, y2, width) 
    plt.bar(x + 3*width, y3, width) 
    plt.xticks(x + 2* width, ('10m/s', '15m/s', '20m/s'))
    
    plt.xlabel('Vehicle speed')
    plt.ylabel('Collision rate')
    plt.legend(labels)
    plt.title(case)
    plt.grid()
```
